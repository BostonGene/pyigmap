```python
import sys
sys.path.append('../bin/cdr3nt_error_corrector/')
import correct, airr
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
```

```python
fname = "SRR1814056.igblast_annotation.tsv.gz"
data = pd.read_csv(fname, sep="\t")
data
```

```python
adata = airr.read_annotation(fname, 
                             only_functional=True, only_canonical=True, 
                             remove_chimeras=True, only_best_alignment=True, 
                             discard_junctions_with_N=True)
adata
```

```python
corr1 = correct.ClonotypeCorrector(top_c_call = False, top_v_alignment_call = False)
corr1.correct_full(adata[0])[["duplicate_count", "v_call", "j_call", "c_call", "junction_aa", "v_sequence_alignment"]]
```

```python
corr2 = correct.ClonotypeCorrector(top_c_call = True, top_v_alignment_call = True)
corr2.correct_full(adata[0])[["duplicate_count", "v_call", "j_call", "c_call", "junction_aa", 
                              "best_v_call", "best_v_sequence_alignment", "best_c_call"]]
```
